```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## Question 1: Use yfinance to Extract Tesla Stock Data

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

## Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
tesla_revenue = pd.read_csv("tesla_revenue.csv")
tesla_revenue.tail()
```

## Question 3: Use yfinance to Extract GameStop Stock Data

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

## Question 4: Use Webscraping to Extract GameStop Revenue Data

```python
gme_revenue = pd.read_csv("gme_revenue.csv")
gme_revenue.tail()
```

## Question 5: Tesla Stock and Revenue Graph

```python
def make_graph(stock_data, revenue_data, stock_name):
    import matplotlib.pyplot as plt

    fig, ax1 = plt.subplots(figsize=(14,6))

    ax1.plot(stock_data['Date'], stock_data['Close'], label='Stock Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price')
    ax1.set_title(f"{stock_name} Stock Price and Revenue")

    ax2 = ax1.twinx()
    revenue_data['Revenue'] = revenue_data['Revenue'].replace('', '0').fillna(0).astype(float)
    ax2.plot(revenue_data['Date'], revenue_data['Revenue'], color='orange', label='Revenue')
    ax2.set_ylabel('Revenue in USD')

    fig.legend(loc='upper left')
    plt.show()
```

```python
make_graph(tesla_data, tesla_revenue, "Tesla")
```

## Question 6: GameStop Stock and Revenue Graph

```python
make_graph(gme_data, gme_revenue, "GameStop")
```
